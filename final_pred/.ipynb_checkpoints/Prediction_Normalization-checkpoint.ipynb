```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import csv
```

```python
subject = '012'
num = '01'
predictions = []
for i in range(1, 6):
    y_true_label = pd.read_csv('edit_subject_'+subject+'_'+num+'__y_prediction_'+str(i)+'.csv', names=['label']).values
    predictions.append(y_true_label)
```

```python
predictions[0].shape[0]
```

```python
norm_predictions = []
for label_index in range(0, predictions[0].shape[0]):
    count = [0,0,0,0]
    for file_index in range(0, 5):
        print(predictions[file_index][label_index][0])
        count[predictions[file_index][label_index][0]] = count[predictions[file_index][label_index][0]] + 1
    #print(count)
    count = np.asarray(count)
    #print('res: ' + str(np.argmax(count)))
    norm_predictions.append(np.argmax(count))
```

```python
norm_predictions = np.asarray(norm_predictions)
```

```python
norm_predictions.shape
```

```python
outfile = open('edit_subject_'+subject+'_'+num+'__y_prediction_6.csv','w')
out = csv.writer(outfile)
out.writerows(map(lambda x: [x], norm_predictions))
outfile.close()
```
